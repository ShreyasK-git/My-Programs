```python
features_val = validation.drop('SalePrice', axis=1)
label_val = validation['SalePrice']
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
def evaluate_model(model, features, labels):
  # Predict on test set
  y_pred = model.predict(features_val)

  # Calculate MAE
  mae = mean_absolute_error(label_val, y_pred)

  # Calculate RMSE
  rmse = np.sqrt(mean_squared_error(label_val, y_pred))

  # Calculate MAPE
  mape = np.mean(np.abs((label_val - y_pred) / label_val)) * 100

  # Print the results
  print(f"MAE: {mae:.4f}")
  print(f"RMSE: {rmse:.4f}")
  print(f"MAPE: {mape:.2f}%")
```
