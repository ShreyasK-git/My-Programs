## Air processing

```python
import json
import pandas as pd

# change the below according to the sample
specimen_file_path = "/content/air_1.bmespecimen"
label = "air"

with open(specimen_file_path, 'r') as f:
    data = json.load(f)

sensor_data = data.get('data')
specimenDataPoints = sensor_data.get('specimenDataPoints')
if specimenDataPoints:
    print("Extracted specimenDataPoints:")
    print(specimenDataPoints)
        # Separate the IDs and data
    ids = [point[0] for point in specimenDataPoints]
    data_values = [point[1:] for point in specimenDataPoints]

    # Create the DataFrame
    df = pd.DataFrame(data_values)

    # Insert the ID column at the beginning
    df.insert(0, 'id', ids)

    print("DataFrame created:")
    display(df.head())
else:
    print("'specimenDataPoints' key not found in the JSON data.")

# getting sensor values for each step in the cycles
dataset = {}
for i in range(0,10):
  filtered_list = df.loc[df[7] == i, 0].tolist()
  dataset[str(i)] = filtered_list
print(dataset)

# convert extracted values to a dataframe
min_len = min(len(v) for v in dataset.values())
for k, v in dataset.items():
    dataset[k] = v[:min_len]
df = pd.DataFrame(dataset)
# df.head()

# Add data label
df['label'] = label
df.head()

# Save the data
df.to_csv(f"{label}.csv", index = False)
```
